Explain the theoretical Simple Linear Regression model in your own words by describing its components (of predictor and outcome variables, slope and intercept coefficients, and an error term) and how they combine to form a sample from normal distribution; then, create python code explicitly demonstrating your explanation using numpy and scipy.stats
<h3>

Simple Linear Regression is a statistical technique utilized to represent the relationship between two variables: a predictor, or independent variable, and an outcome, or dependent variable. The model posits that the result can be forecasted by a linear equation structured as: outcome equals (beta zero) plus (beta one) multiplied by the predictor, plus an error term. In this context, (beta zero), the intercept, signifies the outcome's value when the predictor is zero, while (beta one), the slope, indicates the amount of change in the outcome for each one-unit rise in the predictor. The error term reflects the random variability in the outcome that the predictor cannot solely explain and is generally assumed to follow a normal distribution with a mean of zero. Simple Linear Regression identifies the optimal line that fits through the data points by reducing the gap between the actual points and this line, reflecting the strength and direction of the connection between the predictor and the result.

```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Parameters for the linear model
beta_0 = 2  # Intercept
beta_1 = 1.5  # Slope
sigma = 1  # Standard deviation of the error term

# Generate predictor variable (X) values
np.random.seed(0)
X = np.random.uniform(0, 10, 100)  # Sample 100 points between 0 and 10

# Calculate the expected values of Y based on the model
Y_mean = beta_0 + beta_1 * X

# Generate the outcome variable (Y) by adding normally distributed errors
Y = Y_mean + norm.rvs(scale=sigma, size=X.size)

# Plot the data and the regression line
plt.scatter(X, Y, label="Sampled Data")
plt.plot(X, Y_mean, color="red", label="True Regression Line")
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Simple Linear Regression Model")
plt.legend()
plt.show()
```

<h3>2. Use a dataset simulated from your theoretical Simple Linear Regression model to demonstrate how to create and visualize a fitted Simple Linear Regression model using pandas and import statsmodels.formula.api as smf<h3>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# Step 1: Simulate the Dataset
# Define the parameters for the model
beta_0 = 2      # Intercept
beta_1 = 1.5    # Slope
sigma = 1       # Standard deviation of the error term

# Generate predictor variable (X) values
np.random.seed(0)
X = np.random.uniform(0, 10, 100)  # 100 points between 0 and 10

# Calculate the expected values of Y based on the model
Y_mean = beta_0 + beta_1 * X

# Generate outcome variable (Y) by adding normally distributed errors
Y = Y_mean + np.random.normal(0, sigma, X.size)

# Create a DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Step 2: Fit the Simple Linear Regression Model
# Using statsmodels' formula API to define the model
model = smf.ols('Y ~ X', data=data).fit()

# Print model summary to see coefficients and statistics
print(model.summary())

# Step 3: Visualize the Fitted Model
# Plot the original data points
plt.scatter(data['X'], data['Y'], label="Data Points")

# Plot the fitted regression line
plt.plot(data['X'], model.predict(data), color="red", label="Fitted Regression Line")

# Add labels and title
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Fitted Simple Linear Regression Model")
plt.legend()
plt.show()
```

<h3> 4. Explain how fitted_model.fittedvalues are derived on the basis of fitted_model.summary().tables[1] (or more specifically fitted_model.params or fitted_model.params.values) <h3>

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Simulate some data
np.random.seed(0)
X = np.random.uniform(0, 10, 100)  # 100 data points from a uniform distribution
beta_0 = 2  # Intercept
beta_1 = 1.5  # Slope
sigma = 1  # Standard deviation of the error term

# Generate the outcome variable Y with noise
Y = beta_0 + beta_1 * X + np.random.normal(0, sigma, len(X))

# Create a DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit the model using statsmodels
fitted_model = smf.ols('Y ~ X', data=data).fit()

# Extract fitted values (predicted Y values)
fitted_values = fitted_model.fittedvalues

# Print the fitted values
print("Fitted values (predicted Y values):")
print(fitted_values.head())  # Display the first 5 fitted values

# Visualize the fitted regression line
plt.scatter(data['X'], data['Y'], label="Data Points", color="blue")
plt.plot(data['X'], fitted_values, label="Fitted Regression Line", color="red", linewidth=2)

# Add labels and title
plt.xlabel("Predictor (X)")
plt.ylabel("Outcome (Y)")
plt.title("Simple Linear Regression: Fitted Line vs Data")
plt.legend()
plt.show()
```

The fitted values in simple linear regression are the outcome variable Y's expected values as determined by the fitted regression model. After fitting the regression line to the observed data variance, these fitted values show the estimated link between predictor X and result Y. The predicted coefficients for the intercept and slope are among the parameters that the fitted model gives us when we use statsmodels (or another statistical tool) to fit a linear regression model. The fitted values, or the expected Y values corresponding to each observed value of X based on the regression equation, are calculated using these coefficients.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

# Load the Old Faithful Geyser dataset
old_faithful = sns.load_dataset('geyser')

# Filter for long wait times
long_wait = old_faithful['waiting'] >= 70  # Assuming 70 minutes is considered "long wait time"
long_wait_data = old_faithful[long_wait]

# 1. Bootstrap the fitted slope coefficients
bootstrap_slopes = []
n_bootstrap = 1000

for _ in range(n_bootstrap):
    # Sample with replacement from the long wait time data
    bootstrap_sample = long_wait_data.sample(n=len(long_wait_data), replace=True)
    model = smf.ols('duration ~ waiting', data=bootstrap_sample).fit()
    bootstrap_slopes.append(model.params['waiting'])

# Visualize the bootstrap sampling distribution of the slope coefficients
plt.hist(bootstrap_slopes, bins=30, edgecolor='k', alpha=0.7)
plt.title("Bootstrap Sampling Distribution of the Slope Coefficients")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.show()

# Calculate the 95% confidence interval for the slope coefficient
conf_interval = np.percentile(bootstrap_slopes, [2.5, 97.5])
print(f"95% Bootstrap Confidence Interval for Slope: {conf_interval}")

# 2. Simulate samples under the null hypothesis (1 = 0)
null_slope = 0
intercept = 1.65
sigma = 0.37
n_simulations = 1000
simulated_slopes = []

# Use the existing values of `waiting` to generate `Y` values
for _ in range(n_simulations):
    simulated_Y = intercept + null_slope * long_wait_data['waiting'] + np.random.normal(0, sigma, size=len(long_wait_data))
    simulated_data = pd.DataFrame({'waiting': long_wait_data['waiting'], 'duration': simulated_Y})
    simulated_model = smf.ols('duration ~ waiting', data=simulated_data).fit()
    simulated_slopes.append(simulated_model.params['waiting'])

# Visualize the sampling distribution of the slope under the null hypothesis
plt.hist(simulated_slopes, bins=30, edgecolor='k', alpha=0.7)
plt.title("Sampling Distribution of Slope under Null Hypothesis")
plt.xlabel("Slope Coefficient")
plt.ylabel("Frequency")
plt.show()

# 3. Report if 0 is in the bootstrap CI and compare simulated p-value to original data's p-value
# Check if 0 is within the bootstrap CI
contains_zero = (conf_interval[0] <= 0 <= conf_interval[1])
print(f"Does the 95% confidence interval contain 0? {'Yes' if contains_zero else 'No'}")

# Calculate p-value from the simulated distribution (two-tailed test)
simulated_p_value = np.mean([abs(slope) >= abs(np.mean(bootstrap_slopes)) for slope in simulated_slopes])
print(f"Simulated p-value: {simulated_p_value}")

# Compare with the original model's p-value
original_model = smf.ols('duration ~ waiting', data=long_wait_data).fit()
original_p_value = original_model.pvalues['waiting']
print(f"Original model p-value for slope: {original_p_value}")
```

The "short" and "long" wait times are not "before and after" measurements so there are not natural pairs on which to base differences on which to do a "one sample" (paired differences) hypothesis test; but, we can do "two sample" hypothesis testing using a permuation test, or create a 95% bootstrap confidence interval for the difference in means of the two populations.<h3>

<h4>A) Do a permuation test<h4>

```python
import numpy as np

# Sample data for two groups (you can replace these with actual data)
group_short = np.array([5.1, 4.8, 5.5, 5.3, 5.0])  # wait times for "short" group
group_long = np.array([6.2, 6.5, 6.1, 6.8, 6.3])   # wait times for "long" group

# Number of permutations
n_permutations = 10000

# Step 1: Calculate the observed difference in means
observed_diff = np.mean(group_long) - np.mean(group_short)

# Step 2: Combine both groups into one array
combined_data = np.concatenate([group_short, group_long])

# Step 3: Permutation test
np.random.seed(0)  # for reproducibility
permuted_diffs = []

for _ in range(n_permutations):
    # Shuffle the combined data
    np.random.shuffle(combined_data)
    
    # Split the data into two new groups
    permuted_short = combined_data[:len(group_short)]
    permuted_long = combined_data[len(group_short):]
    
    # Calculate the difference in means for this permutation
    permuted_diff = np.mean(permuted_long) - np.mean(permuted_short)
    permuted_diffs.append(permuted_diff)

# Convert list to numpy array for easier calculation
permuted_diffs = np.array(permuted_diffs)

# Step 4: Calculate the p-value
p_value = np.sum(np.abs(permuted_diffs) >= np.abs(observed_diff)) / n_permutations

print("Observed Difference in Means:", observed_diff)
print("P-value from Permutation Test:", p_value)
```

<h4>(B) Create a 95% bootstrap confidence interval by repeatedly bootstrapping within each group and applying np.quantile(bootstrapped_mean_differences, [0.025, 0.975]) to the collection of differences between the sample means.<h4>

```python
import numpy as np

# Sample data for two groups (you can replace these with actual data)
group_short = np.array([5.1, 4.8, 5.5, 5.3, 5.0])  # wait times for "short" group
group_long = np.array([6.2, 6.5, 6.1, 6.8, 6.3])   # wait times for "long" group

# Number of bootstrap samples
n_bootstrap = 10000

# List to store the bootstrapped differences in means
bootstrapped_mean_differences = []

# Perform bootstrap resampling
np.random.seed(0)  # for reproducibility

for _ in range(n_bootstrap):
    # Resample with replacement within each group
    bootstrap_short = np.random.choice(group_short, size=len(group_short), replace=True)
    bootstrap_long = np.random.choice(group_long, size=len(group_long), replace=True)
    
    # Calculate the mean difference for this bootstrap sample
    mean_diff = np.mean(bootstrap_long) - np.mean(bootstrap_short)
    bootstrapped_mean_differences.append(mean_diff)

# Convert list to numpy array for easier calculation
bootstrapped_mean_differences = np.array(bootstrapped_mean_differences)

# Calculate the 95% confidence interval
confidence_interval = np.quantile(bootstrapped_mean_differences, [0.025, 0.975])

print("95% Bootstrap Confidence Interval for the Difference in Means:", confidence_interval)
```

### (a) Explanation of the Sampling Approaches

1. **Permutation Test**:
   The permutation test is a non-parametric method used to assess the significance of a difference between two groups. It operates by shuffling the group labels (e.g., "short" and "long" wait times) under the assumption that there is no real difference between the groups (null hypothesis). By randomly reassigning the labels and calculating the difference in means for each reshuffling, the test generates a distribution of differences that reflects what we would expect under the null hypothesis. The observed difference in means is then compared to this distribution to calculate a p-value, which tells us the probability of observing a difference as extreme as the one observed, assuming the null hypothesis is true.

2. **Bootstrap Confidence Interval**:
   The bootstrap method involves resampling the data within each group with replacement to create new "bootstrap samples." This process allows us to simulate the sampling distribution of the difference in means between the two groups. By repeatedly calculating the difference in means for these resampled datasets, we generate a distribution of mean differences. From this distribution, we can calculate a confidence interval (typically the 2.5th and 97.5th percentiles) to estimate the range within which the true difference in means is likely to lie, providing an understanding of the uncertainty around the observed difference.

### (b) Comparison with the Indicator Variable-Based Model Approach

The **indicator variable-based model** (like a linear regression) directly includes an indicator variable (typically coded as 0 or 1) to represent group membership. By fitting a regression model, this method estimates the mean difference between the groups by examining the coefficient of the indicator variable. This approach is more parametric and assumes a linear relationship between the predictor (group indicator) and the outcome variable. In addition to estimating the difference, it provides a p-value to test for significance and a confidence interval for the estimated effect, relying on distributional assumptions (such as normality and homogeneity of variance).

In contrast, the **permutation test** and **bootstrap method** are both non-parametric, making fewer assumptions about the data. The permutation test shuffles the data to test the null hypothesis of no difference, while the bootstrap method resamples the data to create a distribution of the difference in means. Both methods provide insights into the significance or uncertainty of the observed difference without assuming any specific distribution of the data, making them particularly useful when data distributions are unknown or non-normal.

The key difference is that the **indicator variable model** is parametric and involves fitting a model to estimate the difference, while the **permutation test** and **bootstrap method** are resampling techniques that generate empirical distributions without relying on specific model assumptions. Both the resampling methods and the indicator variable approach ultimately aim to assess the difference between groups, but they do so through different mechanisms: one via model fitting, and the other via resampling and permutation.
