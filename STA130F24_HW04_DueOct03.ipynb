- **SD** describes the variability of the individual data points in the sample, while **SEM** quantifies the variability of the sample mean.

### 4. **Bootstrapping and SEM:**
- **SD of the sample** refers to the spread of the original data points, while the **SD of bootstrapped means** (after bootstrapping) reflects the **SEM**.

<h3>4. The "Pre-lecture" video (above) mentioned that bootstrap confidence intervals could apply to other statistics of the sample, such as the "median". Work with a ChatBot to create code to produce a 95% bootstrap confidence interval for a population mean based on a sample that you have and comment the code to demonstrate how the code can be changed to produce a 95% bootstrap confidence interval for different population parameter (other than the population mean, such as the population median).<h3>

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Create realistic sample data (e.g., heights in cm)
np.random.seed(0)  # For reproducibility
heights = np.random.normal(loc=175, scale=7, size=100)  # Mean height 175 cm, std dev 7 cm

# Clip the heights to a realistic range (150 cm to 200 cm)
heights = np.clip(heights, 150, 200)

# Display the first few data points
print("Sample Heights Data (first 10 points):", heights[:10])

# Step 2: Function to calculate bootstrap confidence interval for the mean
def bootstrap_ci_mean(data, num_bootstrap=1000, confidence_level=0.95):
    bootstrap_means = np.zeros(num_bootstrap)  # Array to store bootstrap means

    # Generate bootstrap samples and calculate their means
    for i in range(num_bootstrap):
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        bootstrap_means[i] = np.mean(bootstrap_sample)

    # Calculate the lower and upper percentiles for the confidence interval
    lower_bound = np.percentile(bootstrap_means, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(bootstrap_means, (confidence_level + (1 - confidence_level) / 2) * 100)
    
    return bootstrap_means, lower_bound, upper_bound

# Step 3: Calculate the 95% confidence interval for the mean
bootstrap_means, ci_mean_lower, ci_mean_upper = bootstrap_ci_mean(heights, num_bootstrap=1000, confidence_level=0.95)
print(f"95% Bootstrap Confidence Interval for the Mean Height: ({ci_mean_lower}, {ci_mean_upper})")

# Step 4: Function to calculate bootstrap confidence interval for the median
def bootstrap_ci_median(data, num_bootstrap=1000, confidence_level=0.95):
    bootstrap_medians = np.zeros(num_bootstrap)  # Array to store bootstrap medians

    # Generate bootstrap samples and calculate their medians
    for i in range(num_bootstrap):
        bootstrap_sample = np.random.choice(data, size=len(data), replace=True)
        bootstrap_medians[i] = np.median(bootstrap_sample)

    # Calculate the lower and upper percentiles for the confidence interval
    lower_bound = np.percentile(bootstrap_medians, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(bootstrap_medians, (confidence_level + (1 - confidence_level) / 2) * 100)
    
    return bootstrap_medians, lower_bound, upper_bound

# Step 5: Calculate the 95% confidence interval for the median
bootstrap_medians, ci_median_lower, ci_median_upper = bootstrap_ci_median(heights, num_bootstrap=1000, confidence_level=0.95)
print(f"95% Bootstrap Confidence Interval for the Median Height: ({ci_median_lower}, {ci_median_upper})")

# Step 6: Plot histograms of bootstrap means and medians
plt.figure(figsize=(12, 5))

# Histogram for Bootstrap Means
plt.subplot(1, 2, 1)
plt.hist(bootstrap_means, bins=30, color='skyblue', edgecolor='black')
plt.axvline(ci_mean_lower, color='red', linestyle='dashed', linewidth=1)
plt.axvline(ci_mean_upper, color='red', linestyle='dashed', linewidth=1)
plt.title('Bootstrap Means')
plt.xlabel('Mean Height')
plt.ylabel('Frequency')
plt.legend(['95% CI Lower Bound', '95% CI Upper Bound'])

# Histogram for Bootstrap Medians
plt.subplot(1, 2, 2)
plt.hist(bootstrap_medians, bins=30, color='lightgreen', edgecolor='black')
plt.axvline(ci_median_lower, color='red', linestyle='dashed', linewidth=1)
plt.axvline(ci_median_upper, color='red', linestyle='dashed', linewidth=1)
plt.title('Bootstrap Medians')
plt.xlabel('Median Height')
plt.ylabel('Frequency')
plt.legend(['95% CI Lower Bound', '95% CI Upper Bound'])

plt.tight_layout()
plt.show()
```

<h4>1. What is the process of bootstrapping?<h4>

Bootstrapping refers to building or developing something with minimal initial resources. In statistics, bootstrapping is a resampling method used to estimate the distribution of a statistic. The first step in bootstrapping is random sampling with replacement, which means taking a sample of data from the original dataset while allowing the same data point to be selected more than once. The second step involves calculating the statistic for each resampled dataset. This process is repeated multiple times. Finally, the distribution of the statistic is estimated to calculate measures such as variance or confidence intervals.

<h4>2. What is the main purpose of bootstrapping?<h4>


Bootstrapping is a statistical method that helps estimate how reliable a statistic, like an average, is by repeatedly taking random samples from the original data. By resampling with replacement, it allows us to see how much our estimates might vary, even if we only have a small dataset. This approach helps build confidence in our conclusions without needing to make strong assumptions about the data's distribution.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import numpy as np
```

```python
# Load the CSV file
data = pd.read_csv('data.csv')

# Display the first few rows of the DataFrame
print(data.head())

# Display basic statistics
print(data.describe())
```

```python
# Set the style of seaborn
sns.set(style='whitegrid')

# Create a figure with two subplots
plt.figure(figsize=(12, 6))

# Plot Initial Health Scores
plt.subplot(1, 2, 1)
sns.boxplot(x='InitialHealthScore', data=data)
plt.title('Initial Health Scores')

# Plot Final Health Scores
plt.subplot(1, 2, 2)
sns.boxplot(x='FinalHealthScore', data=data)
plt.title('Final Health Scores')

plt.tight_layout()
plt.show()
```

```python
# Calculate mean initial and final health scores
mean_initial = data['InitialHealthScore'].mean()
mean_final = data['FinalHealthScore'].mean()

# Calculate the mean difference
mean_difference = mean_final - mean_initial

print(f'Mean Initial Health Score: {mean_initial}')
print(f'Mean Final Health Score: {mean_final}')
print(f'Mean Difference: {mean_difference}')
```

```python
# Bootstrapping function
def bootstrap_mean_diff(data, n_iterations=1000):
    initial_scores = data['InitialHealthScore']
    final_scores = data['FinalHealthScore']
    
    diffs = []
    for _ in range(n_iterations):
        # Resample with replacement
        sample_initial = initial_scores.sample(frac=1, replace=True).values
        sample_final = final_scores.sample(frac=1, replace=True).values
        diff = sample_final.mean() - sample_initial.mean()
        diffs.append(diff)
        
    return diffs

# Run bootstrap analysis
bootstrapped_diffs = bootstrap_mean_diff(data)

# Calculate confidence intervals
lower_bound = np.percentile(bootstrapped_diffs, 2.5)
upper_bound = np.percentile(bootstrapped_diffs, 97.5)

print(f'Bootstrap 95% Confidence Interval: ({lower_bound}, {upper_bound})')
```

```python
# Conduct a paired t-test
t_statistic, p_value = stats.ttest_rel(data['FinalHealthScore'], data['InitialHealthScore'])

print(f'T-statistic: {t_statistic}')
print(f'P-value: {p_value}')
```

```python
alpha = 0.05  # Significance level

if p_value < alpha:
    conclusion = "Reject the null hypothesis: There is evidence that the vaccine is effective."
else:
    conclusion = "Fail to reject the null hypothesis: There is no evidence that the vaccine is effective."

print(conclusion)
```

**Explanation:**

The reason we used bootstrapping method here is because the original data is limited. We employ bootstrapping to assess the mean of the differences between the initial and final health scores. Prior to using the bootstrapping technique, the differences' mean was 3.299. After utilizing bootstrapping with a confidence level of 95%, we discover that the confidence interval has a lower bound of 1.39 at the 2.5th percentile and an upper bound of 5.40 at the 97.5th percentile. The t-statistic and p-value are used in the code above to help us decide whether to reject the null hypothesis, but they are not always necessary. Only the presence of 0 inside the confidence interval needs to be verified. We can infer that the vaccination affects the patients because 0 is outside of this range. Therefore, we reject the null hypothesis, as there is evidence that the vaccine is effective.
